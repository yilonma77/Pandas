````python
# Installation si nécessaire (décommenter si besoin)
# !pip install anthropic

import anthropic
import os
from IPython.display import Markdown, display
````

## ⚙️ Configuration du client
Renseigne ta clé API Anthropic. Tu peux la stocker dans une variable d'environnement :
```bash
export ANTHROPIC_API_KEY="sk-ant-..."
```

````python
# Clé API — variable d'environnement ou à coller directement
ANTHROPIC_API_KEY = os.environ.get("ANTHROPIC_API_KEY", "sk-ant-...")

client = anthropic.Anthropic(api_key=ANTHROPIC_API_KEY)

# Paramètres du modèle
MODEL       = "claude-3-5-sonnet-20241022"
MAX_TOKENS  = 1024
TEMPERATURE = 1.0

print(f"✅ Client configuré — modèle : {MODEL}")
````

````python
def get_llm_response(prompt: str, system: str = "Tu es un assistant utile et précis.") -> str:
    """
    Envoie un prompt à Claude et retourne la réponse.
    
    Args:
        prompt : La question ou instruction de l'utilisateur
        system : Personnalité / contexte de l'assistant
    
    Returns:
        La réponse textuelle du modèle
    """
    message = client.messages.create(
        model=MODEL,
        max_tokens=MAX_TOKENS,
        temperature=TEMPERATURE,
        system=system,
        messages=[
            {"role": "user", "content": prompt}
        ]
    )
    return message.content[0].text

print("✅ Fonction get_llm_response() prête.")
````

````python
# ✏️ Modifie le prompt et le contexte système ici
prompt = "Explique-moi le théorème de Bayes en 3 phrases simples."

system = "Tu es un assistant pédagogue qui explique des concepts complexes de façon claire et concise."

print(f"📝 Prompt : {prompt}")
````

````python
# 🚀 Appel au LLM et affichage formaté
response = get_llm_response(prompt, system=system)

display(Markdown(f"**Prompt :** {prompt}\n\n---\n\n**Réponse de Claude :**\n\n{response}"))
````

## 💬 ChatGPT (OpenAI)
Même logique avec l'API OpenAI. Clé dispo sur [platform.openai.com/api-keys](https://platform.openai.com/api-keys).
```bash
pip install openai
export OPENAI_API_KEY="sk-..."
```

````python
from openai import OpenAI

# Clé API OpenAI
OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "sk-...")

oai_client = OpenAI(api_key=OPENAI_API_KEY)

# Paramètres du modèle
OAI_MODEL       = "gpt-4o"
OAI_MAX_TOKENS  = 1024
OAI_TEMPERATURE = 1.0

print(f"✅ Client OpenAI configuré — modèle : {OAI_MODEL}")
````

````python
def get_chatgpt_response(prompt: str, system: str = "Tu es un assistant utile et précis.") -> str:
    """
    Envoie un prompt à ChatGPT et retourne la réponse.
    
    Args:
        prompt : La question ou instruction de l'utilisateur
        system : Personnalité / contexte de l'assistant
    
    Returns:
        La réponse textuelle du modèle
    """
    response = oai_client.chat.completions.create(
        model=OAI_MODEL,
        max_tokens=OAI_MAX_TOKENS,
        temperature=OAI_TEMPERATURE,
        messages=[
            {"role": "system", "content": system},
            {"role": "user",   "content": prompt}
        ]
    )
    return response.choices[0].message.content

print("✅ Fonction get_chatgpt_response() prête.")
````

````python
# ✏️ Modifie le prompt ici (peut être le même que Claude pour comparer)
prompt_oai = "Explique-moi le théorème de Bayes en 3 phrases simples."

system_oai = "Tu es un assistant pédagogue qui explique des concepts complexes de façon claire et concise."

print(f"📝 Prompt : {prompt_oai}")
````

````python
# 🚀 Appel à ChatGPT et affichage formaté
response_oai = get_chatgpt_response(prompt_oai, system=system_oai)

display(Markdown(f"**Prompt :** {prompt_oai}\n\n---\n\n**Réponse de ChatGPT ({OAI_MODEL}) :**\n\n{response_oai}"))
````
